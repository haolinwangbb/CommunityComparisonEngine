##Calculated standard error:
The standard error is used to measure the confidence of the sample scale, calculated by the formula: Standard error = sqrt(sample scale x(1-current scale)/ current size)4. Calculate the z-score:
The z-score is calculated using the standard error, which is the sample proportion minus the census proportion divided by the standard error z-score =(sample proportion - census proportion)/ standard error

```python
population_df = population_df.rename(columns={
    'white': 'White alone',
    'black': 'Black or African American alone',
    'asian': 'Asian alone',
    'hispanic': 'Hispanic',
    'other': 'Other'
})
```

```python
import math

standard_errors = pd.DataFrame(index=population_df.index, columns=population_df.columns)

for index, row in population_df.iterrows():
    population_size = sum(row)
    for column in population_df.columns:
        sample_proportion = sample_df[column].values[0]
        standard_error = math.sqrt((sample_proportion * (1 - sample_proportion)) / population_size)
        standard_errors.loc[index, column] = standard_error
```

```python
z_scores = pd.DataFrame(index=population_df.index, columns=population_df.columns)

for index, row in population_df.iterrows():
    for column in population_df.columns:
        population_proportion = row[column]/100
        sample_proportion = sample_df[column].values[0]
        standard_error = standard_errors.loc[index, column]
        z_score = (sample_proportion - population_proportion) / standard_error
        z_scores.loc[index, column] = z_score
```

```python
print("Standard Errors:")
print(standard_errors)
print("\nZ-Scores:")
print(z_scores)
```

##Set the significance level:

```python
import pandas as pd
from scipy import stats

# Set the significance level to 0.05
alpha = 0.05

# Calculate the maximum error probability for accepting the null hypothesis
# Since we are using a one-tailed test, divide the significance level by 2
max_error_probability = alpha / 2

# Check if each sample's Z-score exceeds the significance level
for index, row in z_scores.iterrows():
    for column in z_scores.columns:
        z_score = z_scores.loc[index, column]

        # Look up the critical value in the Z-table (standard normal distribution table)
        # If the calculated Z-score is greater than the critical value, reject the null hypothesis
        # This indicates a significant difference between the category proportion in the sample data and the population data
        if z_score > stats.norm.ppf(1 - max_error_probability):
            print(f"The proportion of {index} category in feature {column} significantly differs from the population proportion.")
        else:
            # If the calculated Z-score is less than or equal to the critical value, fail to reject the null hypothesis
            # This indicates no significant difference between the category proportion in the sample data and the population data
            print(f"The proportion of {index} category in feature {column} does not significantly differ from the population proportion.")
```
