# Comparsion Engine

```python
def comparsion_engine(sample_df, population_df_1, population_df_2,community_district, demographic_feature, alpha):
#sample_df is our sample input data
#community districts can be Queens Community District 1, Manhattan Community District 2, and etc..
#demographic features are only limited to "Ethnicity, Gender, and Age"
#alpha is significance level
    filtered_sample_data = sample_df[sample_df['CD_Name'] == community_district]
    filtered_pop_data = population_df_1[population_df_1['cd_name'] == community_district]
    filtered_age_data = population_df_2[population_df_2['cd_name'] == community_district]
    if demographic_feature == "Ethnicity":
    
        race_proportions = filtered_sample_data['Ethnicity'].value_counts(normalize=True)
        race_N = filtered_sample_data.shape[0]
        
        poplution_df=filtered_pop_data[['pct_white_nh','pct_black_nh','pct_asian_nh','pct_other_nh','pct_hispanic']]
        population_race_pop  = (poplution_df.iloc[0]/100).to_dict()
        
        race_pop = {}
        race_pop['pct_white_nh'] = race_proportions['White']
        race_pop['pct_black_nh'] = race_proportions['Black']
        race_pop['pct_asian_nh'] = race_proportions['Asian']
        race_pop['pct_hispanic'] = race_proportions['Hispanic']
        race_pop['pct_other_nh'] = 1 - sum(race_pop.values())
        
        for race in race_pop.keys():
            print(f"For race {race.replace('pct_', '')}:")


            # Define hypotheses
            print("sample proportion is ", round(race_pop[race], 4), 
                  ", population proportion is ", round(population_race_pop[race], 4))
            if race_pop[race] > population_race_pop[race]:
                print(f"H0: The sample proportion is equal to the population proportion")
                print(f"H1: The sample proportion is greater than the population proportion")
            else:
                print(f"H0: The sample proportion is equal to the population proportion")
                print(f"H1: The sample proportion is less than the population proportion")

            # Calculate test statistic
            z = (race_pop[race] - population_race_pop[race]) / ((population_race_pop[race] * (1 - population_race_pop[race]) / race_N) ** 0.5)
            print(f"Test statistic (Z): {z}")

            if race_pop[race] > population_race_pop[race]:
                # Calculate p-value for the test that sample proportion > population proportion
                p_value = 1 - stats.norm.cdf(abs(z))  # One-tailed test
            else:
                # Calculate p-value for the test that sample proportion < population proportion
                p_value = stats.norm.cdf(-abs(z))  # One-tailed test

            print(f"P-value: {p_value}")

            # Conclusion
            if p_value < alpha:
                if race_pop[race] > population_race_pop[race]:
                    print(f"Conclusion: We reject the null hypothesis. The sample proportion is significantly larger than the population proportion.\n")
                else:
                    print(f"Conclusion: We reject the null hypothesis. The sample proportion is significantly smaller than the population proportion.\n")
            else:
                print(f"Conclusion: We fail to reject the null hypothesis. The sample proportion is not significantly different from the population proportion.\n")
    elif demographic_feature == "Age":
        age_groups = ['Under 5', '5-9', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70-74', '75-79', '80-84', '85+']

        # Create a new dataframe to store the counts
        age_group_counts = pd.DataFrame(columns=['Age Group', 'Count'])

        # Count the occurrences of each age group
        for age_group in age_groups:
            count = filtered_sample_data['Age'].str.contains(age_group).sum()
            percentage = count / len(filtered_sample_data) * 100
            age_group_counts = age_group_counts.append({'Age Group': age_group, 'Count': count, 'Percentage': percentage}, ignore_index=True)
     
        age_groups = ['under_5', '5_9', '10_14', '15_19', '20_24', '25_29', '30_34', '35_39', '40_44', '45_49', '50_54', '55_59', '60_64', '65_69', '70_74', '75_79', '80_84', '85_over']
        # Create a new dataframe to store the summed values
        summed_age_groups = pd.DataFrame(columns=['Age Group', 'Sum'])

        # Iterate over the age groups and sum the corresponding columns
        for age_group in age_groups:
            female_col = f'pop_pct_female_{age_group}'
            male_col = f'pop_pct_male_{age_group}'
            total_sum = filtered_age_data[female_col].sum() + filtered_age_data[male_col].sum()
            summed_age_groups = summed_age_groups.append({'Age Group': age_group.replace('_', '-'), 'Sum': total_sum}, ignore_index=True)
        for i in range(len(age_group_counts)):
            age_group = age_group_counts.loc[i, 'Age Group']
            sample_pct = age_group_counts.loc[i, 'Percentage'] / 100
            pop_pct = summed_age_groups.loc[i, 'Sum'] / summed_age_groups['Sum'].sum()

            print(f"For {age_group}:")
            print(f"Sample proportion: {sample_pct}")
            print(f"Population proportion: {pop_pct}")
            print(f"H0: The sample proportion is equal to the population proportion")
            print(f"H1: The sample proportion is greater than the population proportion")

            # Calculate test statistic
            z = (sample_pct - pop_pct) / ((pop_pct * (1 - pop_pct) / age_group_counts.shape[0]) ** 0.5)
            print(f"Test statistic (Z): {z}")

            # Calculate p-value
            p_value = 1 - stats.norm.cdf(z)  # One-tailed test (greater than)
            print(f"P-value: {p_value}")

            # Conclusion
            if p_value < alpha:
                print("Conclusion: We reject the null hypothesis. The sample proportion is significantly greater than the population proportion.")
            else:
                print("Conclusion: We fail to reject the null hypothesis. The sample proportion is not significantly greater than the population proportion.")

            # Add an extra line break
            print()
    elif demographic_feature == "Gender":
        male_columns = ['pop_pct_male_under_5', 'pop_pct_male_5_9', 'pop_pct_male_10_14',
       'pop_pct_male_15_19', 'pop_pct_male_20_24', 'pop_pct_male_25_29',
       'pop_pct_male_30_34', 'pop_pct_male_35_39', 'pop_pct_male_40_44',
       'pop_pct_male_45_49', 'pop_pct_male_50_54', 'pop_pct_male_55_59',
       'pop_pct_male_60_64', 'pop_pct_male_65_69', 'pop_pct_male_70_74',
       'pop_pct_male_75_79', 'pop_pct_male_80_84', 'pop_pct_male_85_over']  # Add all columns containing male data
        female_columns = ['pop_pct_female_under_5',
               'pop_pct_female_5_9', 'pop_pct_female_10_14', 'pop_pct_female_15_19',
               'pop_pct_female_20_24', 'pop_pct_female_25_29', 'pop_pct_female_30_34',
               'pop_pct_female_35_39', 'pop_pct_female_40_44', 'pop_pct_female_45_49',
               'pop_pct_female_50_54', 'pop_pct_female_55_59', 'pop_pct_female_60_64',
               'pop_pct_female_65_69', 'pop_pct_female_70_74', 'pop_pct_female_75_79',
               'pop_pct_female_80_84', 'pop_pct_female_85_over']  # Add all columns containing female data

        male_percentage = filtered_age_data[male_columns].sum(axis=1)
        female_percentage = filtered_age_data[female_columns].sum(axis=1)
        sample_pct_female = filtered_sample_data[filtered_sample_data['Gender'] == 'Female'].shape[0] / filtered_sample_data.shape[0]
        sample_pct_male = filtered_sample_data[filtered_sample_data['Gender'] == 'Male'].shape[0] / filtered_sample_data.shape[0]
        sample_pct_gender  =  sample_pct_female + sample_pct_male
        sample_pct_female = sample_pct_female / sample_pct_gender
        sample_pct_male =  sample_pct_male / sample_pct_gender
        # Perform the test for each gender
        for gender, sample_pct, pop_pct in [('Female', sample_pct_female, female_percentage.values[0]/100), ('Male', sample_pct_male, male_percentage.values[0]/100)]:
            print(f"For {gender}:")
            print(f"Sample proportion: {sample_pct}")
            print(f"Population proportion: {pop_pct}")
            print(f"H0: The sample proportion is equal to the population proportion")
            print(f"H1: The sample proportion is not equal to the population proportion")

            # Calculate test statistic
            z = (sample_pct - pop_pct) / ((pop_pct * (1 - pop_pct) / filtered_sample_data.shape[0]) ** 0.5)
            print(f"Test statistic (Z): {z}")

            # Calculate p-value
            p_value = 2 * stats.norm.cdf(-abs(z))  # Two-tailed test
            print(f"P-value: {p_value}")

            # Conclusion
            if p_value < alpha:
                print("Conclusion: We reject the null hypothesis. The sample proportion is significantly different from the population proportion.\n")
            else:
                print("Conclusion: We fail to reject the null hypothesis. The sample proportion is not significantly different from the population proportion.\n")

```

```python
#If we reject the null hypothesis, it indicates that our survey is not representative in certain groups, 
#such as different age groups, race groups, or gender groups. In other words, 
#the survey results may not accurately reflect the opinions or characteristics of these specific groups. 
#On the other hand, if we fail to reject the null hypothesis, 
#it suggests that the survey we conducted is well distributed across these groups. 
#This means that the survey results are more likely to be representative and statistically significant, 
#indicating that any observed patterns or relationships in the data are not due to chance alone.
```
